```python
node1_count = new_phis.groupby("node1").agg({"node2": "count"})
node1_count.rename({"node2": "degree"}, axis="columns", inplace=True)
node2_count = new_phis.groupby("node2").agg({"node1": "count"})
node2_count.rename({"node1": "degree"}, axis="columns", inplace=True)
print(node1_count.head())
print(node2_count.head())
combined = pd.concat([node1_count, node2_count])
combined = combined.reset_index().groupby("index").sum()
new_phis = new_phis.merge(combined, left_on="node1", right_on="index", how="left")
new_phis = new_phis.merge(combined, left_on="node2", right_on="index", how="left", suffixes=["1", "2"])
print(new_phis.head())
```

```python
new_phis_filtered = new_phis[(new_phis["degree1"] < 50) & (new_phis["degree2"] < 50)]
print(new_phis_filtered.head(20))
```
